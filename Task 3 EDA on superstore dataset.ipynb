### heatmap of data

```python
plt.figure(figsize=(7,4))
sns.heatmap(data.corr(), cmap = 'viridis', annot = True, linewidth = 4)
#annot = True: write the data value in each cell
```

# Ship Mode wise Sales Analysis

```python
data['Ship Mode'].value_counts()
```

```python
# Used a countplot for Ship Mode
plt.figure(figsize=(6,4))
sns.countplot(x=data["Ship Mode"])
plt.show()
```

# Region wise Analysis

```python
data.columns
```

```python
data['Region'].value_counts()
```

```python
#using Barplot to show profits across different regions
plt.figure(figsize = (6,4))
sns.set(font_scale=1)
sns.barplot(x='Region', y='Profit', data=data, hue='Segment')
plt.title("Profit across different Regions(by Segments)", fontsize = 15)
plt.show()
```

```python
# Using pie chart to show region sales
data['Region'].value_counts().plot(kind="pie",figsize=(5,5),autopct="%1.2f%%",shadow="True")
plt.legend(fontsize=10)
plt.show
```

```python
#using Barplot to show profits across different regions
plt.figure(figsize = (6,4))
sns.set(font_scale=1)
sns.barplot(x='Region', y='Sales', data=data, hue='Segment')
plt.title("Sales across different Regions(by Segments)", fontsize = 15)
plt.show()
```

```python
#using Barplot to show profits across different regions
plt.figure(figsize = (6,4))
sns.set(font_scale=1)
sns.barplot(x='Region', y='Discount', data=data, hue='Segment')
plt.title("Discount given across different Regions(by Segments)", fontsize = 15)
plt.show()
```

# Region wise Ship Mode Analysis

```python
plt.figure(figsize=(7,6))
sns.countplot(x="Region", hue="Ship Mode", data=data,palette='magma')
plt.ylabel("Ship Mode Count")
plt.show()
```

* Observation:- Standard Class is most opted Ship Mode and Same Day is least opted Ship Mode among different regions.

# Sales and Profit distribution among different Regions.

```python
sp = data.groupby("Region")[["Sales","Profit"]].sum().sort_values(by=["Sales","Profit"],ascending=False)
sp.plot(kind="bar", fontsize=10, color=["purple","pink"])
plt.xlabel("Region", fontsize=10)
plt.ylabel("Sales and Profit", fontsize=10)
plt.show()
```

# City Wise Analysis

```python
data['City'].value_counts()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x=data['City'], order=(data['City'].value_counts().head(60)).index)
plt.xticks(rotation=60)
plt.xlabel('City', fontsize = 15)
plt.ylabel('Count / No.of buyers', fontsize = 15)
plt.show()
```

# StateWise Sales Analysis

```python
data['State'].value_counts()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x=data['State'], order=(data['State'].value_counts().head(60)).index)
plt.xticks(rotation=60)
plt.xlabel('States', fontsize = 15)
plt.ylabel('Count / No.of buyers', fontsize = 15)
plt.show()
```

* Observation: California, New York, Texas has highest count of buyers and Wyoming has the least count

# CategoryWise Sales Analysis

```python
data['Category'].value_counts()
```

```python
sns.barplot(x=data['Category'],y=data['Sales'],palette='magma')
plt.show()
```

Observation: Sales are mostly in Technology and Furniture Category

```python
#Barplot for profits corresponding to each category
sns.barplot(x='Category', y='Profit', data=data, palette="magma")
plt.show()
```

Observation:There is more profit in Technology Category and moderate in Furniture and Office Supplies category.

```python
cat_sale_profit = data.groupby("Category")[["Sales","Profit"]].sum().sort_values(by=["Sales","Profit"], ascending = True)
cat_sale_profit.plot(kind="bar",fontsize=10, color = ["green", "red"])
plt.xlabel("Category", fontsize=10)
plt.ylabel("Sales and Profit", fontsize=10)

plt.show()
```

furniture sale is good but the not making profit

# Sales and Profit distribution among sub-category

```python
sls = data.groupby("Sub-Category")[["Sales","Profit"]].sum().sort_values(by=["Profit"], ascending = False)
sls.plot(kind="bar", figsize=(8,5), fontsize=10, color = ["orange", "purple"])
plt.xticks(rotation=70)
plt.xlabel("Sub-Category", fontsize=10)
plt.ylabel("Sales and Profit", fontsize=10)
plt.show()
```

Observation: Papers shows highest profit with moderate sales and Tables shows least profit with higher sales.
